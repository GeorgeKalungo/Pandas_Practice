```python
# read data
titanic = pd.read_csv(r'Datasets\original_titanic.csv')
titanic.head()
```

```python
# Set the multi-index for the data with the columns 'sex' and 'name'.
titanic_index = titanic.set_index(['sex', 'name']).head()
titanic_index
```

```python
# Extract the columns ['sex', 'name', 'survived', 'pclass'] as two separate datasets and combine them using the join method.
data1 = titanic[['sex', 'name', 'survived']]
data2 = titanic[['pclass', 'survived', 'sex']]

print('data1:\n', data1.head())
print('data2:\n', data2.head())
```

```python
titanic_join = data1.join(data2, lsuffix='_data_1', rsuffix='_data_2')
titanic_join.head()
```

```python

 #Concat the two datasets given below with the level names as 'data1' and 'data2':
    # - data1 = titanic[['sex', 'name', 'survived']]
    # - data2 = titanic[['pclass', 'age', 'cabin']]
 
data1 = titanic[['sex', 'name', 'survived']]
data2 = titanic[['pclass', 'age', 'cabin']]

print('data1:\n', data1.head())
print('data2:\n', data2.head())
```

```python
pd.concat([data1, data2], keys=['data1', 'data2'], join='outer')
```

```python
pd.concat([data1, data2], keys=['data1', 'data2'], join='inner', sort=True)
```

```python
# Stack the titanic data on the columns 'sex' and 'name' and observe how the data looks.
titanic_reindex = titanic.set_index(['sex', 'name'])
titanic_reindex.head()
```
