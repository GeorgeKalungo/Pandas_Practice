#### Exercise_02:
_For the data read in exercise_01:_
 1. _Describe the data using pandas usual method and guess the percentage of survived with age._
 2. _Display the data with respect to sex, and count the number of male and female passengers in the ship._
 3. _Find the total number of passengers whose age is greater than or equal to 50 and also less than 50._
 4. _Count the total number of passengers who survived in the risk, and the total number of male and female passengers who survived._
 5. _Export the data into a modified1_titanic.csv by removing the columns ['name', 'parch', 'ticket', 'cabin', 'boat', 'body', 'home.dest']._

```python
print(data.head(10))
```

```python
# 1(a) Describe the data
data.describe()
```

```python
# 1(b) Guess the percentage of survived with age
survived_ser = data.loc[:, 'age'].groupby(data['survived'] == True).count()
no_of_passengers_survived = survived_ser.iloc[1].sum()
total_no_of_passengers = data.loc[:, 'survived'].count()
percentge_survived = no_of_passengers_survived/ total_no_of_passengers
print(percentge_survived)
```

```python
# 2(a) display the data with respect to sex
# data.loc[:, 'sex'].head(10)
male = data[data['sex'] == 'male']
female = data[data['sex'] == 'female']

print(male)
print()
print(female)
```

```python
male_count = male['sex'].count()
female_count = female['sex'].count()
female_count

print('Males:', male_count)
print('Females:', female_count)
```

```python
# 3(a) No.of passengers whose age is >= 50
len(data[data['age'] >= 50])
```

```python
# 3(b) No.of passengers whose age is < 50
len(data[data['age'] < 50])
```

```python
# 4(a) The total number of passengers who survived in the risk
survived_series = data['survived'].value_counts()
count_survived = survived_series.iloc[1]
count_survived
```

```python
survive_percent = count_survived/ (survived_series.iloc[0] + survived_series.iloc[1])
survive_percent
```

```python
# 4(b) Total number of males and females who survived
male_survivors = len(data[(data['survived'] == 1) & (data['sex'] == 'male')])
female_survivors = len(data[(data['survived'] == 1) & (data['sex'] == 'female')])
print('Male Survivors:', male_survivors)
print('Female Survivors:', female_survivors)
```

```python
# 5. Export the data into a csv file
data.columns.values
data.to_csv(r'Datasets\modified1_titanic.csv', index=False, columns= ['survived', 'pclass', 'sex', 'age', 'sibsp', 'fare', 'embarked', 'deck'])
```
