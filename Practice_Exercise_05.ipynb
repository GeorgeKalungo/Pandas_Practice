#### Exercise01:
##### For the titanic data set in the directory data/titanic:
 1. _Display the data with the 'name' column as the index._
 2. _Count the number of null and non-null values in the column 'body' using isnull and notnull methods._
 3. _Drop the rows and columns which are having all the null values._
 4. _Replace all the null values of titanic data with 0._


```python
# Display the data with the 'name' column as the index
df = pd.read_csv(r'Datasets\original_titanic.csv', index_col= 'name')
df.head()
```

```python
# Count the number of null and non-null values in the column 'body' using isnull and notnull methods
null_values = int(df['body'].isnull().sum())
non_null_values = int(df['body'].notnull().sum())
print(f'Null Values: {null_values}')
print(f'Non-Null Values: {non_null_values}')
```

```python
# Drop the rows and columns which are having all the null values
df.dropna(how='all', axis=0, inplace=True)
df.shape
```

```python
df.dropna(how='all', axis=1, inplace=True)
df.shape
```

```python
# Replace all the null values of titanic data with 0
df.fillna(0, inplace=True)
```

 1. _Count the number of rows which are duplicated and display them._

 3. _Count the number of 'tickets' with repeated number and display them._
 4. _Convert the 'sex' column names 'male' and 'female' to a title case._
 5. _Replace the 'male' and 'female' names with the title case 'Male' and 'Female'._
 6. _Rename the column names of titanic data to title case._
 7. _Find the number of passengers with different age groups and rename the age groups with different names._

```python
# Count the number of rows which are duplicated and display them
no_of_duplicate_rows = int(df.duplicated().sum())
print(f'Number of duplicated rows: {no_of_duplicate_rows}')
```

```python
# Drop the duplicated rows from the titanic data
df.drop_duplicates(inplace=True)
```

```python
# Count the number of 'tickets' with repeated number and display them
no_of_duplicated_tickets = int(df['ticket'].duplicated().sum())
print(f'Number of duplicated "ticket" rows: {no_of_duplicated_tickets}')
```

```python
# Convert the 'sex' column names 'male' and 'female' to a title case
f = lambda x: x.title()
df['sex'] = df['sex'].apply(f)
df.head(3)
```

```python
# Replace the 'male' and 'female' names with the title case 'Male' and 'Female'
df['Sex'].replace(to_replace= ['male', 'female'], value=['Male', 'Female'])
df.head(3)
```

```python
# Rename the column names of titanic data to title case
df.rename(columns= str.title, inplace=True)
df.head(3)
```

```python
# Find the number of passengers with different age groups and rename the age groups with different names
age_series = pd.Series(df['Age'])
age_groups = pd.qcut(age_series, 3, precision=0, labels= ['Minors', 'Young Adults', 'Adults'])
```

```python
# convert the age_groups into a categorical object
age_groups = pd.Categorical(age_groups)
age_groups.value_counts()
```

```python
# Count the number of passengers with age > 50
age_gt_50 = df['Age'] > 50
no_of_passengers_age_gt_50 = int(df[age_gt_50]['Age'].count())
print(f'No. of Passengers with age > 50: {no_of_passengers_age_gt_50}')
```
