##### Exercise01:
_For the titanic data:_
1. _Group the titanic data based on the 'sex' column and find the mean age of the male and female passengers._
2. _Use the indexing method to find the mean age instead of the groupby method._
3. _Group the data based on the 'sex' column and find the number of male and female passengers who survived._
4. _Count the number of passengers that survived and those who didn't based on the gender and cross-check the result._
5. _Group the data based on the 'sex' column and iterate over the key and data._



```python
# read the titanic data
titanic = pd.read_csv(r'Datasets\original_titanic.csv')
titanic.head()
```

```python
# 1. Group the titanic data based on the 'sex' column and find the mean age of the male and female passengers
titanic_grouped = titanic['age'].groupby(by=titanic['sex']).mean()
titanic_grouped
```

```python
# 2. Use the indexing method to find the mean age instead of the groupby method
titanic['age'][titanic['sex'] == 'male'].mean().round(2)
```

```python
titanic['age'][titanic['sex'] == 'female'].mean().round(2)
```

```python
# 3.Group the data based on the 'sex' column and find the number of male and female passengers who survived
t_gsex = titanic['survived'].groupby(by=titanic['sex'])
t_gsex.sum()


```

```python
# 4. Count the number of passengers that survived and those who didn't based on the gender and cross-check the result
titanic_grouped_survived = titanic[titanic['survived'] == 1].groupby(by='sex')['survived'].count()
titanic_grouped_not_survived = titanic[titanic['survived'] == 0].groupby(by='sex')['survived'].count()
print('Survived')
print(titanic_grouped_survived)
print()
print('Died')
print(titanic_grouped_not_survived)
```

```python
titanic.groupby(by=['sex', 'survived']).size()
```

```python
# 5. Group the data based on the 'sex' column and iterate over the key and data
titanic_grouped_i = titanic.groupby(by='sex')
for (key, data) in titanic_grouped_i:
    print(key)
    print(data)
```
