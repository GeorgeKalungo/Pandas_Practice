##### _Exercise_01_
- _For the dataframe shown below:_
  1. _Sort the DataFrame by 'name' in descending order and then by attempts in ascending order._
  2. _Drop the 'attempts' column from the DataFrame._
  3. _Reindex the columns as 'name', 'attempts', 'score' and 'qualify'._
  4. _Reindex the row labels as 'A', 'C', 'B', 'D', ..., 'J'._
  

```python
import pandas as pd
import numpy as np
```

```python
exam_data = {
    'name': ['Anjun', 'Anali', 'Souria', 'Rockea', 'Imani', 'Mouliya', 'Julie', 'Rana', 'Kavin', 'Bosia'],
    'score': [15.5, 9, 16.5, np.nan, 9, 30, 17.5, np.nan, 8, 20],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
```

```python
data = pd.DataFrame(exam_data, index= labels)
print(data)
```

```python
# 1(a) sorting the data by name in descending order
data.sort_values(by='name', ascending=False)
```

```python
# 1(b) sorting the data by attempts in ascending order
data.sort_values(by='attempts')
```

```python
# dropping the attempts column
data.drop(columns= 'attempts')
```

```python
# 3. reindexing the columns
new_col_indx = ['name', 'attempts', 'score', 'qualify']
# data.reindex(columns= new_col_indx)
data.reindex(new_col_indx, axis='columns')
```

```python
# 4. reindexing the row labels
new_row_labels = ['A', 'C', 'B', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
data.reindex(index= new_row_labels)
```

#### _Exercise_02_
##### _For the DataFrame created in exercise_01:_
 1. _Find the maximum score and the maximum attempts made by the students using the apply function._
 2. _Find the maximum, minimum, and mean of the score and the attempts made by the students using a user defined function._
 3. _Sort the index labels in reverse direction._
 4. _Sort the DataFrame based on the values of score and attempts._

```python
print(data)
data.loc[:, ['score', 'attempts']]
```

```python
# 1(a) max values in score and attempts columns
f = lambda x: x.max()
data.loc[:, ['score', 'attempts']].apply(f)
```

```python
# 1(b) min values in score and attempts columns
f = lambda x: x.min()
data.loc[:, ['score', 'attempts']].apply(f)
```

```python
# 2. User defined function
def f(x):
    return pd.Series([x.max(), x.min(), x.mean()], index= ['max', 'min', 'mean'])

data.loc[:, ['score', 'attempts']].apply(f)
```

```python
# 3. sorting the index labels in reverse
print(data)
# data in reverse direction index-wise
print()
data.sort_index(ascending=False)
```

```python
# 4. sorting the dataframe based on the score and attempts values
print(data)
# sort by score column
print()
print('Sorting by the "score" column:')
print(data.sort_values(by= 'score'))

# sort by the attempts column
print()
print('Sorting by the "attempts" column:')
print(data.sort_values(by= 'attempts'))
```
